```python
import numpy as np
import pandas as pd

data = pd.read_csv('./data/discovery-park-clean.csv', na_values="*", index_col=0)
head = data.head(100).copy()
head
```

```python
data['smoothed_alt'] = data['gps_altitude_m'].ewm(span=10, adjust=False).mean()

from matplotlib import pyplot as plt
#plt.plot(head[['smoothed_alt', 'gps_altitude_m']])
```

```python

data[['latitude','longitude','smoothed_alt']].to_numpy()
```

```python
import simplekml
kml = simplekml.Kml()

simplekml.LineString()
path = kml.newlinestring(name='Actual path')
path.coords = data[['longitude','latitude','smoothed_alt']].to_numpy()
path.extrude = 1
path.altitudemode = simplekml.AltitudeMode.absolute

kml.save('./data/discovery-park.kml')
```
